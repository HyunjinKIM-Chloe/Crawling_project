```python
import pandas as pd
import numpy as np
path = "../../datas/"
```

```python
weight_df = pd.read_excel(path+"cat_weights.xlsx")
weight_df
```

```python
wdf = weight_df[['category', 'colname','weights']]
weight_dict = wdf.to_dict('records')
```

```python
import pickle
with open(path+'weight_dict.txt', 'wb') as wd:
    pickle.dump(weight_dict, wd)
```

```python
with open(path+'weight_dict.txt', 'rb') as wd:
    weight_dict = pickle.load(wd)
    print(weight_dict)
```
